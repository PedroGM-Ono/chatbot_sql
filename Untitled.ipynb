````python

import pandas as pd
import sqlite3
from sqlalchemy import (
    create_engine,
    MetaData,
    Table,
    Column,
    String,
    Integer,
    Float,
    insert,
    inspect,
    text,
)
db_path = "database/vendas_retail.db"
engine = create_engine(f"sqlite:///{db_path}")
inspector = inspect(engine)
````

````python
tables = inspector.get_table_names()

description="Dado as seguintes tabelas:"

inspector = inspect(engine)
for table in inspector.get_table_names():
    columns_info = [(col["name"], col["type"]) for col in inspector.get_columns(table)]

    table_description = f"Tabela {table}:\n"

    table_description += "Colunas:\n" + "\n".join([f"  {name}: {col_type}" for name, col_type in columns_info])
    description += "\n\n" + table_description

description += "\nMe retorne apenas o texto da query, pronto para ser executado"
description += "\n\nMe escreva uma query em SQL para:\n"

print(description)
````

````python
# Executa a consulta no SQLite e retorna o resultado como DataFrame
def executar_query(sql):
    try:
        conn = sqlite3.connect(DB_PATH)
        df = pd.read_sql_query(sql, conn)
        conn.close()
        return df
    except Exception as e:
        print("Erro ao executar SQL:", e)
        return None
````
